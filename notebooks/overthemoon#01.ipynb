## <데이터 가공2>

- 새로운 데이터 프레임 (missions)만들기

```python
missions=pd.DataFrame()
missions
```

- Mission의 고유값 찾기

```python
rock_samples['Mission'].unique()
```

- missions에 새로운 컬럼 추가

```python
missions['Mission']=rock_samples['Mission'].unique()
missions
```

```python
missions.info()
```

- 아폴로 임무별 샘플 중량 총합

```python
sample_total_weight=round(rock_samples.groupby('Mission')['Weight (kg)'].sum(),3)
sample_total_weight
```

- 데이터프레임과 시리즈 연결하기

```python
missions=missions.merge(sample_total_weight,on='Mission')
missions
```

- 컬럼명 변경하기 (Weight (kg==>Sample weight (kg)))

```python
missions.rename(columns={'Weight (kg)':'Sample weight (kg)'},inplace=True)
missions
```

- 임무간 중량 차이구하기

```python
missions['Sample weight (kg)'].diff()
```

- 컬럼명 생성하여 diff값 추가하기

```python
missions['Weight diff']=missions['Sample weight (kg)'].diff()
missions
```

- 결측치 채우기 : 0으로 채우기

```python
missions.fillna(0,inplace=True)
missions
```

## <데이터 가공3>

- 달탐사선 중량 데이터 추가
    - 달모듈 + 명령모듈 =승무원모듈

- 달 모듈(Lunar Module)

```python
# 달 모듈 이름과 중량 추가
missions['Lunar module (LM)']=['Eagle (LM-5)','Interpid (LM-6)','Antares (LM-8)','Falcol (LM-10)','Orion (LM-11)','Challenger (LM-12)'] #달 모듈이름
missions['LM mass (kg)']=[15103,15235,15264,16430,16445,16456] # 달모듈중량
missions
```

- 임무간 달 모듈 중량 차이 구하기

```python
missions['LM mass diff']=missions['LM mass (kg)'].diff()
missions.fillna(value=0,inplace=True)
missions
```

- 명령모듈(command module)

```python
# 명령 모듈이름과 중량추가
missions['command module (CM)']=['columbia (CSM-107)', 'Yankee Clipper (CM-108)','Kitty hawk (CM-110)','Endeavor (CM-112)','Casper (CM-113)','America (CM-114)']
missions['CM mass (kg)']=[5560,5609,5758,5875,5840,5960]

```

- 임무간 명령 모듈 중량 차이 구하기

```python
missions['CM mass diff']=missions['CM mass (kg)'].diff()
missions.fillna(value=0,inplace=True)
missions
```

- 승무원영역 중량 데이터 추가

```python
# 승무원영역 =  달모듈+명령모듈

missions['Total weight (kg)']=missions['LM mass (kg)'] + missions['CM mass (kg)']
missions['Total weight diff']=missions['LM mass diff']+missions['CM mass diff']
missions
```

- 페이로드에서 승무원이 차지하는 비율

```python
saturnVPayload=43500

missions['Crewed area : Payload']=missions['Total weight (kg)']/saturnVPayload
```

- 샘플이 차지하는 비율구하기

```python
# 승무원영역에서 샘플이 차지하는 비율
missions['Sample : Crewedarea']=missions['Sample weight (kg)']/missions['Total weight (kg)']
# 페이로드에서 샘플이 차지하는 비율
missions['Sample : Payload']=missions['Sample weight (kg)']/saturnVPayload
missions
```
