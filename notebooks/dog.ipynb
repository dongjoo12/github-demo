```python
import numpy as np
import matplotlib.pyplot as pit
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 기존 데이터 닥스훈트와 진돗개
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]
 
jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

#새로운데이터
newdata_lengh = [59]
newdata_height = [35]
```

```python
pit.scatter(dach_length, dach_height, label='dachshund')
pit.scatter(jin_length, jin_height, label='jindo')
pit.scatter(newdata_lengh, newdata_height, marker='p', label='new')
pit.legend(loc='upper right')
pit.show()
```

```python
#데이터 테이블로 만들기
d_data =np.column_stack((dach_length, dach_height))
d_label = np.zeros(len(d_data))

j_data= np.column_stack((jin_length, dach_height))
j_label = np.ones(len(d_data))

dogs = np.concatenate((d_data, j_data))
lables = np.concatenate((d_label, j_label))
dogs, lables
```

```python
k=3
knn = KNeighborsClassifier(n_neighbors=k)
knn
```

```python
knn.fit(dogs, lables)
```

```python
newdata = np.column_stack((newdata_lengh, newdata_height))
```

```python
y_pred = knn.predict(newdata)
y_pred
```

```python
dog_classes = {0:'닥스훈트', 1: '진돗개'}
print(f'예측데이터 : {newdata}, 예측결과 : {y_pred[0]}')
```
